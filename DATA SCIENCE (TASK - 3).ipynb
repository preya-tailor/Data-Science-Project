```python
import pandas as pd

df= pd.read_csv('Downloads/advertising.csv').fillna(0)
dataframe = pd.DataFrame(df)
print(dataframe)
```

```python
dataframe.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 10))

sns.scatterplot(data=dataframe, x='TV', y='Sales',s=50)
plt.xlabel('RADIO')
plt.ylabel('SALES')
plt.title('TV Advertising AND Sales')
```

```python
sns.scatterplot(data=dataframe, x='Radio', y='Sales',s=50)
plt.xlabel('TV')
plt.ylabel('SALES')
plt.title('Radio vs Sales')
```

```python
sns.scatterplot(data=dataframe, x='Newspaper', y='Sales',s=50)
plt.xlabel('NEWSPAPER')
plt.ylabel('SALES')
plt.title('Newspaper vs Sales')
```

```python
from sklearn.model_selection import train_test_split

x = dataframe.iloc[:,0:3].values
y = dataframe.iloc[:,3].values

xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error, mean_squared_error

regr = LinearRegression()
regr.fit(xtrain, ytrain)
# ypred = regr.predict(xtest)
ypred = regr.predict(xtest)

r2score = r2_score(ytest,ypred)
mae = mean_absolute_error(y_true=ytest, y_pred=ypred)
mse = mean_squared_error(y_true=ytest, y_pred=ypred)
rmse = mean_squared_error(y_true=ytest, y_pred=ypred, squared=False)

print('R2SCORE: ', r2score)
print('MAE: ', mae)
print('MSE: ', mse)
print('RMSE: ', rmse)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = sns.regplot(x=ytest,y=ypred, color='blue')
plt.xlabel('ACTUAL VALUES')
plt.ylabel('PREDICTED VALUES')
plt.show(fig)
```
